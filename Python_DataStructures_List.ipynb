```python
# second largest number in list
list1 =  [12, 45, 98, 34, 23]
large_number = max(list1)
sec_large = list1[0]
for i in list1:
    if i != large_number and i > sec_large:
        sec_large = i
print(sec_large)
```

```python
#Write a Python script that takes a list of numbers and returns the sum of all numbers, but exclude any duplicates.
lists = [1,2,3,2,4,5,2,6,3,5,2]
total_sum = 0
new_list = set(lists)
for i in new_list:
    total_sum += i
print(total_sum)
```

```python
# Find the sum of ONLY the numbers that appeared more than once
lists = [1,2,3,2,4,5,2,6,3,5,2]
dic = {}
total_sum = 0
for num in lists:
    if num in dic:
        dic[num] += 1
    else:
        dic[num] = 1
print(dic)
for num in dic:
    if dic[num ]>1:
        total_sum+= num
print("total sum",total_sum)
```

```python
# Find even and odd numbers in a list 
list1 = [12, 17, 19, 24, 35, 40, 56]
even = []
odd = []
for num in list1:
    if num%2 == 0:
        even.append(num)
    else:
        odd.append(num)
print("even numbers: ",even)
print("odd numbers: ",odd)
```

```python
# You have a list of temperatures in Celsius. Create a new list that converts them to Fahrenheit.
celsius = [0, 10, 20, 30, 40]
fahrenheit = []
for i in celsius:
    F = ((i* 9/5) + 32)
    fahrenheit.append(F)
print(fahrenheit)
```

```python
# List comprehension method
celsius = [0, 10, 20, 30, 40]

# Standard List Comprehension: [expression for item in list]
fahrenheit = [(i * 9/5) + 32 for i in celsius]

print(fahrenheit)
```
