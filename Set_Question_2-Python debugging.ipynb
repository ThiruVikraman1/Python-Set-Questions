```python
#2.Find the error in the following Python code:
def my_function():
    x += 5
    print(x)

x = 10
my_function()
```

```python
# corrected code
def my_function():
    global x      # x should be declared a global 
    x += 5        # x is modified in the function
    print(x)

x = 10
my_function()
```
